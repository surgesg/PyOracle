```python
import pyoracle
from IPython.core.display import Image
```

```python
filename = 'audio/prokofiev.wav'
fft_size = 8192
hop_size = 8192
```

```python
features = pyoracle.make_features(filename, fft_size, hop_size)
```

```python
oracle = pyoracle.make_oracle(0.09, features, 'mfcc', 2)
```

```python
pyoracle.draw_oracle(oracle, 'output.png', size=(640*8, 480*8))
Image(filename='output.png')
```

```python
ir = pyoracle.calculate_ir(oracle, 1.15)
plot(ir[0][0], ir[0][1])
xlim((0, 122))
ylabel("IR", fontsize=18)
xlabel("Oracle State", fontsize=18)
title("Information Rate of Musical Example", fontsize=18)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12, 8)
fig.savefig('IR_plot.png', dpi=600)
```
