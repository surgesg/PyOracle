```python
import pyoracle
```

```python
filename = 'mmp_proj/audio/GH_COMP_no_silence-141rep.wav'
fftsize = 8192
hopsize = 8192
```

```python
features = pyoracle.make_features(filename, fftsize, hopsize)
```

```python
range = (0.005, 0.1, 0.025)
n_frames = int(round(44100.0 / fftsize)) # used to get 1Hz oracle rate
ideal_t = pyoracle.calculate_ideal_threshold(range, features, 'mfcc', n_frames)
print n_frames
mfcc_oracle = pyoracle.make_oracle(ideal_t[1], features, 'mfcc', n_frames)
```

```python
range = (0.005, 0.1, 0.025)
n_frames = int(round(44100.0 / fftsize)) # used to get 1Hz oracle rate
ideal_t = pyoracle.calculate_ideal_threshold(range, features, 'chroma', n_frames)
print n_frames
chroma_oracle = pyoracle.make_oracle(ideal_t[1], features, 'chroma', n_frames)
```

```python
gh_mfcc_ir = array(pyoracle.calculate_ir(mfcc_oracle)[0])
gh_mfcc_ir = gh_mfcc_ir / max(gh_mfcc_ir)
plot(gh_mfcc_ir)
print len(gh_mfcc_ir)
```

```python
gh_chroma_ir = array(pyoracle.calculate_ir(chroma_oracle)[0])
gh_chroma_ir = gh_chroma_ir / max(gh_chroma_ir)
plot(gh_chroma_ir)
print len(gh_chroma_ir)
```
