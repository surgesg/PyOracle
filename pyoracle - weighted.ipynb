Extract features from audio file - uses Bregman suite.

```python
features = pyoracle.make_features('audio/prokofiev.wav', 8192)
```

```python
num_frames = 2
chroma_oracle = pyoracle.make_oracle(0.01, features, 'chroma', num_frames)
mfcc_oracle = pyoracle.make_oracle(0.01, features, 'mfcc', num_frames)
rms_oracle = pyoracle.make_oracle(0.01, features, 'rms', num_frames)
centroid_oracle = pyoracle.make_oracle(0.01, features, 'centroid', num_frames)
zerocrossings_oracle = pyoracle.make_oracle(0.01, features, 'zerocrossings', num_frames)
```

```python
mfcc_ir, code, compror = pyoracle.calculate_ir(mfcc_oracle) 
chroma_ir, code, compror = pyoracle.calculate_ir(chroma_oracle) 
centroid_ir, code, compror = pyoracle.calculate_ir(centroid_oracle) 
rms_ir, code, compror = pyoracle.calculate_ir(rms_oracle) 
zerocrossings_ir, code, compror = pyoracle.calculate_ir(zerocrossings_oracle) 
```

```python
weights_vec = []
num_states = len(mfcc_oracle)

for i in range(num_states):
    current_weights = {'mfcc': 0, 'centroid': 0, 'chroma':0, 'rms':0, 'zerocrossings':0} 

    compare = {}
    
    compare['mfcc'] = mfcc_ir[i]
    compare['chroma'] = chroma_ir[i]
    compare['centroid'] = centroid_ir[i]
    compare['rms'] = rms_ir[i]
    compare['zerocrossings'] = zerocrossings_ir[i]
    
    compare = sorted(compare.iteritems(), key=lambda x:x[1], reverse=True)
    
    current_weights[compare[1][0]] = 1
    
    weights_vec.append(current_weights)

    
```

```python
weights_vec
```

```python
meta_oracle = pyoracle.make_dynamic_oracle(0.1, features, weights_vec, num_frames) 
```

```python
pyoracle.draw_oracle(meta_oracle, 'metaoutput.png', size=(640, 480))
Image(filename='metaoutput.png')
```
